### 8. In this exercise, we will generate simulated data, and will then use this data to perform best subset selection.

#### a) Use the rnorm() function to generate a predictor X of length n = 100, as well as a noise vector ε of length n = 100.


```r
set.seed(1)
X = rnorm(100)
eps = rnorm(100)
```

#### b) Generate a response vector Y of length n = 100 according to the model

$Y = \beta_0 + \beta_1 X + \beta_2 X^2 + \beta_3 X^3 + \epsilon$,

#### where $\beta_0, \beta_1, \beta_2$ and $\beta_3$ are constants of your choice


```r
beta0 = 2
beta1 = 1
beta2 = 0.3
beta3 =  5
Y = beta0 + beta1 * X + beta2 * X^2 + beta3 * X^3 + eps
```

#### c)  Use the regsubsets( ) function to perform best subset selection in order to choose the best model containing the predictors $X$, $X^2$ , . . . , $X^{10}$ . What is the best model obtained according to C$_p$ , BIC, and adjusted $R^2$ ? Show some plots to provide evidence for your answer, and report the coefficients of the best model ob- tained. Note you will need to use the data.frame() function to create a single data set containing both $X$ and $Y$.

```r
library(leaps)
df = data.frame(y = Y, x = X)

#Poly hace un polinomio de grado dos
reg.fit = regsubsets(y ~ poly(x, 10, raw = T), data = df, nvmax = 10)
Summary = summary(reg.fit)

# Find the model size for best cp, BIC and adjr2
which.min(Summary$cp)
```

```r
#The which.min() function can be used to identify the location of the minumum point of a vector.
which.min(Summary$bic)
```

```r
#The which.max() function can be used to identify the location of the maximum point of a vector.
which.max(Summary$adjr2)
```

```r
# Plot cp, BIC y adjr2
plot(Summary$cp, xlab = "Número de variables", ylab = "Cp", pch = 20, type = "l")
points(3, Summary$cp[3], pch = 4, col = "red", lwd = 7)
```

```r
plot(Summary$bic, xlab = "Número de variables", ylab = "BIC", pch = 20, type = "l")
points(3, Summary$bic[3], pch = 4, col = "red", lwd = 7)
```

```r
plot(Summary$adjr2, xlab = "Número de variables", ylab = "Adjusted R2", pch = 20, type = "l")
points(4, Summary$adjr2[4], pch = 4, col = "red", lwd = 7)
```

### 10. We have seen that as the number of features used in a model increases, the training error will necessarily decrease, but the test error may not. We will now explore this in a simulated data set.

#### a)Generate a data set with p = 20 features, n = 1,000 observations, and an associated quantitative response vector generated according to the model

$Y=X\beta + \epsilon$

```r
set.seed(1)
p=20
n=1000
x = matrix(rnorm(n * p),n,p) #Creamos la matriz de n p
beta = rnorm(p) # beta debe tener algunos parametros =0
beta[2]  =0
beta[10] =0 
beta[4]  =0
beta[15] =0
epsilon = rnorm(1000)
y = x %*% beta + epsilon #Creamos la función
```

#### b) Split your dataset into a trainingset containing 100 observations and a test set containing 900 observations.

```r
entrenamiento = sample(seq(n), 100, replace = FALSE) #Tomamos una muestra de 100, con novecientos
y.entrenamiento = y[entrenamiento, ]
y.test = y[-entrenamiento, ]
x.entrenamiento = x[entrenamiento, ]
x.test = x[-entrenamiento, ]
```

#### c) Perform best subset selection on the training set, and plot the training set MSE associated with the best model of each size.

```r
library(ISLR)
library(leaps)
regfit.full = regsubsets(y ~ ., data = data.frame(x = x.entrenamiento, y = y.entrenamiento), nvmax = p)
val.errors = rep(NA, p)
x_cols = colnames(x, do.NULL = FALSE, prefix = "x.")
for (i in 1:p) {
    coefi = coef(regfit.full, id = i)
    pred = as.matrix(x.entrenamiento[, x_cols %in% names(coefi)]) %*% coefi[names(coefi) %in% x_cols]
    val.errors[i] = mean((y.entrenamiento - pred)^2)
}
plot(val.errors, ylab = "Training MSE", pch = 19, type = "b")
```

```r
val.errors = rep(NA, p)
for (i in 1:p) {
    coefi = coef(regfit.full, id = i)
    pred = as.matrix(x.test[, x_cols %in% names(coefi)]) %*% coefi[names(coefi) %in% x_cols]
    val.errors[i] = mean((y.test - pred)^2)
}
plot(val.errors, ylab = "Test MSE", pch = 19, type = "b")
```
